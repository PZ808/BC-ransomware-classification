```python
dft = pd.read_csv('dft.csv')
```

```python
dft.n_day.min()
```

```python
# initialize mapper
mapper = km.KeplerMapper(verbose=1)
```

```python
dft.label.value_counts()
```

```python
dft_CW = dft[(dft['label']=='CryptoWall')|(dft['label']=='CryptoLocker')]
```

```python
graph_fs = ['length', 'weight', 'count', 'neighbors','looped',
            'income']
```

```python
dft_CW=dft_CW[graph_fs].reset_index(drop=True)
```

```python
dft_CW
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(dft_CW)
```

```python
dft_CW = scaler.transform(dft_CW)
```

```python
dft_CW
```

```python
# project data into 2D subsapce via 2 step transformation, 1)isomap 2)UMAP
projected_data = mapper.fit_transform(dft_CW, projection=[manifold.Isomap(n_components=6, n_jobs=-1), umap.UMAP(n_components=2,random_state=1)])
```

```python

G = mapper.map(projected_data, dft_CW, clusterer=sklearn.cluster.DBSCAN(metric="cosine"))
```

```python
# define an excessively long filename (helpful if saving multiple Mapper variants for single dataset)
fileID = 'projection=' + G['meta_data']['projection'].split('(')[0] + '_' + \
'n_cubes=' + str(G['meta_data']['n_cubes']) + '_' + \
'perc_overlap=' + str(G['meta_data']['perc_overlap']) + '_' + \
'clusterer=' + G['meta_data']['clusterer'].split('(')[0] + '_' + \
'scaler=' + G['meta_data']['scaler'].split('(')[0]
```

```python
# visualize graph
mapper.visualize(G,
                path_html="mapper_example_" + fileID + ".html",
                title=fileID,
             #   custom_tooltips = np.arange(len(graph_fs)),
                #color_values = dft_CW['length'].to_numpy(),
                color_function_name = 'Length',
                node_color_function = np.array(['average', 'std', 'sum', 'max', 'min']))

# display mapper in jupyter
km.jupyter.display("mapper_example_2" + fileID + ".html")
```
